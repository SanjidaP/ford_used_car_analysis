# Data Visualization

#### Import required libraries and packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ford.csv")
df.head()
```

### About data Set:
1. This is a Ford carcompany dataset 
2. This data set have total 17964 Row and 8 Columns
3. Among them there are 3 column consists catagorical data.

#### barplots side by side

```python
sns.set_context('paper')

plt.figure(figsize=(30,20))

plt.subplot(2,2,1)
model_count = df['model'].value_counts()
model_count = pd.DataFrame(model_count)
model_count = model_count.reset_index()
model_count = model_count.rename(columns={'index':'model', 'model':'No_of_cars'})
model_count['% of cars'] = np.round(((model_count['No_of_cars']/model_count['No_of_cars'].sum())*100), 2)
sns.barplot(x='model', y='% of cars', data=model_count, color='blue', alpha=0.75)
plt.title("Fig 1(a): Number of cars present for each model")
plt.xlabel("Ford car models")
plt.ylabel("Percent of cars (%)")
plt.xticks(rotation=90)
plt.yticks(np.arange(0,41,10).tolist())


plt.subplot(2,2,2)
transmission_count = df['transmission'].value_counts()
transmission_count = pd.DataFrame(transmission_count)
transmission_count = transmission_count.reset_index()
transmission_count = transmission_count.rename(columns={'index':'transmission', 'transmission':'No_of_cars'})
transmission_count['% of cars'] = np.round(((transmission_count['No_of_cars']/transmission_count['No_of_cars'].sum())*100), 2)
sns.barplot(x='transmission', y='% of cars', data=transmission_count, color='orange')
plt.title("Fig 1(b): Number of cars present for each transmission type")
plt.xlabel("Type of transmission")
plt.ylabel("Percent of cars (%)")
plt.yticks(np.arange(0,101,10).tolist())


plt.subplot(2,2,3)
fuelType_count = df['fuelType'].value_counts()
fuelType_count = pd.DataFrame(fuelType_count)
fuelType_count = fuelType_count.reset_index()
fuelType_count = fuelType_count.rename(columns={'index':'fuel_type', 'fuelType':'No_of_cars'})
fuelType_count['% of cars'] = np.round(((fuelType_count['No_of_cars']/fuelType_count['No_of_cars'].sum())*100), 2)
sns.barplot(x='fuel_type', y='% of cars', data=fuelType_count, color='green', alpha=0.75)
plt.title("Fig 1(c): Number of cars present for each fuelType")
plt.xlabel("Type of fuel")
plt.ylabel("Percent of cars (%)")
plt.yticks(np.arange(0,71,10).tolist())


plt.suptitle("Categorical variables in the Ford used cars dataset")
plt.show()
plt.tight_layout()
```

### About bar plot:
1. Fig 1(a) shows that mostly ford produces is Fiesta. From graph, we can see that more than 35% of the car is Fiesta.
2. Fig 1(b) shows that most of the car Ford produces is manual transmission type.
3. Fig 1(b) shows that very few of the car Ford produces is semi-auto transmission type.
4. Fig 1(c) shows that 65% of the car Ford produces runs by petrol.

#### Screeplot

```python
plt.figure(figsize=(20,12))

# adding a categorical feature: 'transmission'
plt.subplot(1,2,1)
sns.scatterplot(x='mileage', y='price', data=df, hue='transmission')
plt.title("Fig 2(a): transmisson type data on milage vs price plot")

# adding a categorical feature: 'fuelType'
plt.subplot(1,2,2)
sns.scatterplot(x='mileage', y='price', data=df, hue='fuelType')
plt.title("Fig 2(b): fuleType type data on milage vs price plot")


plt.suptitle("Numarical variables in the Ford used cars dataset")
plt.show()
plt.tight_layout()
plt.show()
```

### About Scree plot:
1. Fig 2(a) shows that they mostly produces manual cars and they are available in all price ranges.
2. Fig 2(a) shows most of the car produced have fuel type petrol or diesel 

#### Seaborn histogram plot

```python
plt.figure(figsize=(40,35))

plt.subplot(3,2,1)
sns.histplot(data=df, x='price', binwidth=10000, hue= "transmission", element="step", common_norm=False)
plt.title("Fig 3(a): Number of car produced on basis of price between range 10K (basis of transmission)")

plt.subplot(3,2,2)
sns.histplot(data=df, x='price', binwidth=5000)
plt.title("Fig 3(b): Number of car produced on basis of price between range 1K")


plt.subplot(3,2,3)
sns.histplot(data=df, x='year', binwidth=2, hue= "transmission")
plt.title("Fig 3(c): Car production count each 5 year interval")

plt.subplot(3,2,4)
sns.histplot(data=df, x='mpg', binwidth=5)
plt.title("Fig 3(d): Number of car prcuction on basis of mpg")


plt.subplot(3,2,5)
sns.histplot(data=df, x='engineSize', binwidth=0.3)
plt.title("Fig 3(e): Number of car prcuction on basis of EngineSize")

plt.suptitle("Fig 3: Seaborn Histogram of numarical variables in the Ford used cars dataset")
plt.show()
plt.tight_layout()
plt.show()
```

### About Seaborn Hist plot:
1. Fig 3(a) shows that cars Ford produces are mostly Manual transmission types.
2. Fig 3(b) shows that most of the Ford car price range is from 5000 to 15000 
3. Fig 3(c) shows that Ford company's number of car production has an increasing trend. And they are pronto produce Manual transmission type car.
4. Most of the cars Ford company produces provides mileage from 50 to 70 mpg
5. Ford company produces mostly 1 unit size engine type cars

#### Distribution plot

```python
sns.set_context('talk')

plt.figure(figsize=(40,30))

plt.subplot(2,2,1)
sns.histplot(data=df, x='price', binwidth=12000, kde=True)
plt.title("Fig 4(a): Number of car produced on basis of price between range 12K (basis of transmission)")


plt.subplot(2,2,2)
sns.histplot(data=df, x='year', binwidth=5,  kde=True )
plt.title("Fig 4(b): Car production count each 5 year interval")

plt.subplot(2,2,3)
sns.histplot(data=df, x='mpg', binwidth=5, kde=True)
plt.title("Fig 4(c): Number of car prcuction on basis of mpg")


plt.subplot(2,2,4)
sns.histplot(data=df, x='engineSize', binwidth=.5, kde=True)
plt.title("Fig 4(d): Number of car prcuction on basis of EngineSize")

plt.suptitle("Fig 4: Seaborn data distribution of numarical variables in the Ford used cars dataset")
plt.show()
plt.tight_layout()
plt.show()
```

### About Distribution plot:
1. The distrbution generated in figure 4(a) is unimodal.
2. Distribution generated in figure 4(b) is a multimodal.
3. Distribution generated in figure 4(c) is a multimodal.
4. Distribution generated in figure 4(d) is a multimodal.

#### Cumulative plot

```python
sns.set_context('talk')

plt.figure(figsize=(40,30))

plt.subplot(2,2,1)
sns.histplot(data=df, x='price', binwidth=1200, cumulative=True, stat='probability',  hue='transmission')
plt.grid()
plt.title("Fig 5(a): Cumulative plot of car produced on basis of price between range 12K (basis of fuelType)")



plt.subplot(2,2,2)
sns.histplot(data=df, x='year', binwidth=1, hue= "fuelType", cumulative=True)
plt.title("Fig 5(b): Cumulative car production count each 5 year interval")

plt.subplot(2,2,3)
sns.histplot(data=df, x='mpg', binwidth=10, cumulative=True, stat='probability')
plt.title("Fig 3(c): Number of car prcuction on basis of mpg")


plt.suptitle("Fig 3: Seaborn Histogram of numarical variables in the Ford used cars dataset")
plt.show()
plt.tight_layout()
plt.show()
```

### About cumulative data plot:
1. Fig 3(a) shows that if a ford car price is 20000 or less there is a 80% chance the car is Manual transmission type.
2. Fig 3(b) shows that their car production has increasing trend and most of the car uses petrol as fuel
3. Fig 3(c) shows that almost 65% of the Ford car provises milage less than 60.
4. Fig 3(c) shows that very few Ford car povides milage more than 75.
