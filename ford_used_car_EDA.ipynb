# Data Visualization

#### Import required libraries and packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ford.csv")
df.head()
```

### About data Set:
1. This is a Ford carcompany dataset 
2. This data set have total 17964 Row and 8 Columns
3. Among them there are 3 column consists catagorical data.

#### barplots side by side

```python
sns.set_context('paper')

plt.figure(figsize=(30,20))

plt.subplot(2,2,1)
model_count = df['model'].value_counts()
model_count = pd.DataFrame(model_count)
model_count = model_count.reset_index()
model_count = model_count.rename(columns={'index':'model', 'model':'No_of_cars'})
model_count['% of cars'] = np.round(((model_count['No_of_cars']/model_count['No_of_cars'].sum())*100), 2)
sns.barplot(x='model', y='% of cars', data=model_count, color='blue', alpha=0.75)
plt.title("Fig 1(a): Number of cars present for each model")
plt.xlabel("Ford car models")
plt.ylabel("Percent of cars (%)")
plt.xticks(rotation=90)
plt.yticks(np.arange(0,41,10).tolist())


plt.subplot(2,2,2)
transmission_count = df['transmission'].value_counts()
transmission_count = pd.DataFrame(transmission_count)
transmission_count = transmission_count.reset_index()
transmission_count = transmission_count.rename(columns={'index':'transmission', 'transmission':'No_of_cars'})
transmission_count['% of cars'] = np.round(((transmission_count['No_of_cars']/transmission_count['No_of_cars'].sum())*100), 2)
sns.barplot(x='transmission', y='% of cars', data=transmission_count, color='orange')
plt.title("Fig 1(b): Number of cars present for each transmission type")
plt.xlabel("Type of transmission")
plt.ylabel("Percent of cars (%)")
plt.yticks(np.arange(0,101,10).tolist())


plt.subplot(2,2,3)
fuelType_count = df['fuelType'].value_counts()
fuelType_count = pd.DataFrame(fuelType_count)
fuelType_count = fuelType_count.reset_index()
fuelType_count = fuelType_count.rename(columns={'index':'fuel_type', 'fuelType':'No_of_cars'})
fuelType_count['% of cars'] = np.round(((fuelType_count['No_of_cars']/fuelType_count['No_of_cars'].sum())*100), 2)
sns.barplot(x='fuel_type', y='% of cars', data=fuelType_count, color='green', alpha=0.75)
plt.title("Fig 1(c): Number of cars present for each fuelType")
plt.xlabel("Type of fuel")
plt.ylabel("Percent of cars (%)")
plt.yticks(np.arange(0,71,10).tolist())


plt.suptitle("Categorical variables in the Ford used cars dataset")
plt.show()
plt.tight_layout()
```

### About bar plot:
1. Fig 1(a) shows that mostly ford produces is Fiesta. From graph, we can see that more than 35% of the car is Fiesta.
2. Fig 1(b) shows that most of the car Ford produces is manual transmission type.
3. Fig 1(b) shows that very few of the car Ford produces is semi-auto transmission type.
4. Fig 1(c) shows that 65% of the car Ford produces runs by petrol.

#### Boxplots of all numeric features

```python
def separate_num_cols(dataframe):

    # create an empty list to keep the numeric columns
    num_cols = []

    # create a for loop to check the data types of all columns except year column
    for col in dataframe.columns:
        if (col !='year') & (dataframe[col].dtypes == 'int64') | (dataframe[col].dtypes == 'float64'):
            num_cols.append(col)
        
    return num_cols

numeric_cols = separate_num_cols(df)

plt.figure(figsize=(20,15))
for index in range(len(numeric_cols)):
    plt.subplot(2,3,index+1)
    sns.boxplot(x=df[numeric_cols[index]])

plt.suptitle("Boxplot of the numerical variables in the telecom dataset")
plt.show()
plt.tight_layout()
plt.show()
```

### About box plot of Ford car
1. Price boxplot shows, price range is of Ford car is 0 to 25000 and median value near 11000.
2. Mileage boxplot of Ford car dataset shows that it has a range between 0 to 600000. Median value is near 25000
3. Ford car mpg range 25 to abour 80 mpg. Median value is near 60.
4. Median eingine size of Ford car is 1.
